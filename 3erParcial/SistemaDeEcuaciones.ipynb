```python
#   Sistema de ecuaciones
from matplotlib import pyplot as plt

error_reg = []
error_reg.append(1)
x1_reg = []
x2_reg = []
x3_reg = []
x4_reg = []
x5_reg = []

def gn(v,x1,x2,x3,x4,x5):
    if v == 1:
        return (0.25 + 0.25*x2)
    elif v == 2:
        return (0.25 + 0.25*x2 + 0.25*x3)
    elif v == 3:
        return (0.25 + 0.25*x2 + 0.25*x4)
    elif v == 4:
        return (0.25 + 0.25*x3)
    elif v == 5:
        return (0)

def convergencia():
    parada = input("Criterio de paro: ")
    if(parada == ""):
        parada = 0.0001

    n = int(input("Numero de incognitas que posee tu sistema de ecuaciones (maximo 5): "))
    a = input("Si tus valores inciales son iguales a 0 presiona ENTER de lo contrario ingresa cualquier letra y luego presiona ENTER:")

    if n <= 5 and a == "":
            x1_reg.append(0)
            x2_reg.append(0)
            x3_reg.append(0)
            x4_reg.append(0)
            x5_reg.append(0)
    elif n <= 5:
        for i in range(n):
            print("valor de x",i+1)
            if i == 0:
                x1_reg.append(float(input()))
            elif i == 1:
                x2_reg.append(float(input()))
            elif i == 2:
                x3_reg.append(float(input()))
            elif i == 3:
                x4_reg.append(float(input()))
            elif i == 4:
                x5_reg.append(float(input()))
    # si la lista no se usa aun asi se registrara un cero en esta
        if not x3_reg: x3_reg.append(0)
        if not x4_reg: x4_reg.append(0)
        if not x5_reg: x5_reg.append(0)
    else: print("rango superado")

    contador = 0
    v = 1
    x1 = 0
    x2 = 0
    x3 = 0
    x4 = 0
    x5 = 0
    while(contador <= 500):
        '''
        if(float(error_reg[contador]) <= float(parada)):
            return contador
        '''
        if v > n:
            v = 1
            contador += 1
            
        x1 = float(x1_reg[contador])
        x2 = float(x2_reg[contador])
        x3 = float(x3_reg[contador])
        x4 = float(x4_reg[contador])
        #x5 = float(x5_reg[contador])

        x=gn(v,x1,x2,x3,x4,x5)

        if v == 1:
            x1_reg.append(x)
        elif v == 2:
            x2_reg.append(x)
        elif v == 3:
            x3_reg.append(x)
        elif v == 4:
            x4_reg.append(x)
        elif v == 5:
            x5_reg.append(x)

        '''error_reg.append(abs(xn_reg[contador+1] - xn_reg[contador]))'''
        v += 1
    return contador

def mostrar(contador):
    if(contador < 500):
        print("Converge")
    else:
        print("No converge")

    for registro in x1_reg:
        print("x1=", registro)
    for registro in x2_reg:
        print("x2=", registro)
    for registro in x3_reg:
        print("x3=", registro)
    for registro in x4_reg:
        print("x4=", registro)
    for registro in x5_reg:
        print("x5=", registro)

    for registro in error_reg:
        print("Error:", registro)

    plt.plot(x1_reg, label='x1')
    plt.plot(x2_reg, label='x2')
    plt.plot(x3_reg, label='x3')
    plt.plot(x4_reg, label='x4')
    plt.plot(x5_reg, label='x5')

    plt.xlabel('Eje x')
    plt.ylabel('Eje y')
    plt.title("Valores de x")
    plt.legend()
    plt.show()

    plt.plot(error_reg)
    plt.xlabel('Eje x')
    plt.ylabel('Eje y')
    plt.title("Error")
    plt.show()

mostrar(convergencia())
```
